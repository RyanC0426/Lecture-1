Step 4: Data Cleaning and Visualization

```python
df1 = df.iloc[:,[0,1,7,8,9,10,11]]

import scipy.stats as stats
import matplotlib.pyplot as plt
column_title = ['FARM','YEAR','YIT','X1','X2','X3','X4']
plt.figure(figsize = (12,12))
for i in range(len(column_title)):
  plt.subplot(4,2,i+1)
  stats.probplot(df1[column_title[i]], dist = "norm", plot = plt)

plt.tight_layout(pad = 1.0, w_pad = 2.5, h_pad = 5.0)
```

According to the q-q plot, we notice that the YIT,X1,X2,X4 are normally distributed, since they fit the lowess line very well.
